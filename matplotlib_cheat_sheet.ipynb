```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
X = np.linspace(0, 4 * np.pi, 1000)
Y = np.sin(X)
fig, ax = plt.subplots()
ax.plot(X, Y)
fig.show()
```

```python
X = np.linspace(-10, 10, 10)
Y = X * X
fig, ax = plt.subplots()
ax.set_title('Title')
ax.set_ylabel('y') 
ax.set_xlabel('x')
ax.plot(X, Y, color='black', linestyle='--', linewidth=3, marker='o')
fig.show()
```

```python
X = np.random.uniform(0, 1, 100) 
Y = np.random.uniform(0, 1, 100) 
fig, ax = plt.subplots()
ax.scatter(X, Y)
fig.show()
```

```python
X = np.arange(10) 
Y = np.random.uniform(1, 10, 10)
fig, ax = plt.subplots()
ax.bar(X, Y)
fig.show()
```

```python
X = np.arange(10) 
Y = np.random.uniform(1, 100, 10)
fig, ax = plt.subplots()
ax.bar(X, Y)

X = np.linspace(0, 10, 10)
Y = X * X
ax.plot(X, Y, color='red')
fig.show()
```

```python
X = np.random.uniform(0, 1, 4)
fig, ax = plt.subplots()
ax.pie(X)
fig.show()
```

```python
X = np.random.normal(0, 1, 100)
fig, ax = plt.subplots()
ax.hist(X)
fig.show()
```

```python
img = mpimg.imread('/content/ship.png')
fig, ax = plt.subplots()
ax.imshow(img)
fig.show()
```

```python
X = np.linspace(0, 4 * np.pi, 1000) 
Y1, Y2 = np.sin(X), np.cos(X) 
fig, ax = plt.subplots()
ax.plot(X, Y1, X, Y2)
fig.show()
```

```python
X = np.linspace(0, 4 * np.pi, 1000) 
Y1, Y2 = np.sin(X), np.cos(X) 
fig, (ax1, ax2) = plt.subplots(2, 1) 
ax1.plot(X, Y1, color='C1')
ax2.plot(X, Y2, color='C0')
fig.show()
```

```python
X = np.linspace(0, 4 * np.pi, 1000) 
Y1, Y2 = np.sin(X), np.cos(X) 
fig, (ax1, ax2) = plt.subplots(1, 2) 
ax1.plot(X, Y1, color='C1')
ax2.plot(X, Y2, color='C0')
fig.show()
```
